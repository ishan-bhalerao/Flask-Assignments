from flask import Flask, render_template, request, redirect, url_for, session

app = Flask(__name__)
app.secret_key = 'your_secret_key'  # Change this to a random, secret key

@app.route('/', methods=['GET', 'POST'])

    if request.method == 'POST':
        user_input = request.form['user_input']

        return redirect(url_for('display_data'))
    return render_template('form.html')

    user_data = session.get('user_data', 'No data available')

# 7. Integrate a SQLite database with Flask to perform CRUD operations on a list of items.

<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Flask SQLite CRUD</title>
</head>
<body>
    <h1>Items</h1>
    <ul>
        {% for item in items %}
            <li>
                {{ item[1] }}
                <a href="{{ url_for('delete_item', item_id=item[0]) }}">Delete</a>
            </li>
        {% endfor %}
    </ul>
    <form action="{{ url_for('add_item') }}" method="post">
        <label for="name">Item Name:</label>
        <input type="text" name="name" required>
        <button type="submit">Add Item</button>
    </form>
</body>
</html>

from flask import Flask, render_template, request, redirect, url_for
import sqlite3

app = Flask(__name__)

# SQLite database configuration
DATABASE = 'your_database.db'

def create_table():
    conn = sqlite3.connect(DATABASE)
    cursor = conn.cursor()
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS items (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            name TEXT NOT NULL
        )
    ''')
    conn.commit()
    conn.close()

@app.before_first_request
def before_first_request():
    create_table()

@app.route('/')
def index():
    conn = sqlite3.connect(DATABASE)
    cursor = conn.cursor()
    cursor.execute('SELECT * FROM items')
    items = cursor.fetchall()
    conn.close()
    return render_template('index.html', items=items)

@app.route('/add', methods=['POST'])
def add_item():
    if request.method == 'POST':
        name = request.form['name']
        conn = sqlite3.connect(DATABASE)
        cursor = conn.cursor()
        cursor.execute('INSERT INTO items (name) VALUES (?)', (name,))
        conn.commit()
        conn.close()
    return redirect(url_for('index'))

@app.route('/delete/<int:item_id>')
def delete_item(item_id):
    conn = sqlite3.connect(DATABASE)
    cursor = conn.cursor()
    cursor.execute('DELETE FROM items WHERE id = ?', (item_id,))
    conn.commit()
    conn.close()
    return redirect(url_for('index'))

if __name__ == '__main__':
    app.run(debug=True)

from flask import Flask, render_template, redirect, url_for, request, flash
from flask_login import LoginManager, UserMixin, login_user, login_required, logout_user, current_user
import sqlite3

app = Flask(__name__)
app.config['SECRET_KEY'] = 'your_secret_key'

# SQLite database configuration
DATABASE = 'your_database.db'

login_manager = LoginManager(app)
login_manager.login_view = 'login'

class User(UserMixin):
    pass

@login_manager.user_loader
def load_user(user_id):
    user = User()
    user.id = user_id
    return user

def create_users_table():
    conn = sqlite3.connect(DATABASE)
    cursor = conn.cursor()
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS users (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            username TEXT NOT NULL,
            password TEXT NOT NULL
        )
    ''')
    conn.commit()
    conn.close()

@app.before_first_request
def before_first_request():
    create_users_table()

@app.route('/')
def index():
    return 'Welcome to the Flask Login App!'

@app.route('/login', methods=['GET', 'POST'])
def login():
    if request.method == 'POST':
        username = request.form['username']
        password = request.form['password']

        conn = sqlite3.connect(DATABASE)
        cursor = conn.cursor()
        cursor.execute('SELECT * FROM users WHERE username = ?', (username,))
        user_data = cursor.fetchone()
        conn.close()

        if user_data and password == user_data[2]:
            user = User()
            user.id = user_data[0]
            login_user(user)
            flash('Login successful!', 'success')
            return redirect(url_for('dashboard'))
        else:
            flash('Login failed. Please check your username and password.', 'error')

    return render_template('login.html')

@app.route('/logout')
@login_required
def logout():
    logout_user()
    flash('Logout successful!', 'success')
    return redirect(url_for('login'))

@app.route('/dashboard')
@login_required
def dashboard():
    return f'Hello, {current_user.id}! You are logged in.'

if __name__ == '__main__':
    app.run(debug=True)

#9. Create a RESTful API using Flask to perform CRUD operations on resources like books or movies.

from flask import Flask, request
from flask_restful import Resource, Api

app = Flask(__name__)
api = Api(app)

# In-memory database for demonstration purposes
books = [
    {'id': 1, 'title': 'Book 1', 'author': 'Author 1'},
    {'id': 2, 'title': 'Book 2', 'author': 'Author 2'},
    # Add more books as needed
]

class BookResource(Resource):
    def get(self, book_id=None):
        if book_id:
            book = next((item for item in books if item['id'] == book_id), None)
            if book:
                return {'data': book}
            else:
                return {'message': 'Book not found'}, 404
        else:
            return {'data': books}

    def post(self):
        new_book = {
            'id': max(b['id'] for b in books) + 1,
            'title': request.json['title'],
            'author': request.json['author']
        }
        books.append(new_book)
        return {'data': new_book}, 201

    def put(self, book_id):
        book = next((item for item in books if item['id'] == book_id), None)
        if book:
            book['title'] = request.json['title']
            book['author'] = request.json['author']
            return {'data': book}
        else:
            return {'message': 'Book not found'}, 404

    def delete(self, book_id):
        global books
        books = [item for item in books if item['id'] != book_id]
        return {'message': 'Book deleted successfully'}

api.add_resource(BookResource, '/books', '/books/<int:book_id>')

if __name__ == '__main__':
    app.run(debug=True)

# 12 Build a Flask app that updates data in real-time using WebSocket connections.

from flask import Flask, render_template
from flask_socketio import SocketIO, emit
import threading
import time

app = Flask(__name__)
socketio = SocketIO(app)

counter_value = 0
thread = None
thread_lock = threading.Lock()

def update_counter():
    global counter_value
    while True:
        time.sleep(1)
        with thread_lock:
            counter_value += 1
            socketio.emit('update', {'value': counter_value}, namespace='/test')

@app.route('/')
def index():
    return render_template('index_realtime.html')

@socketio.on('connect', namespace='/test')
def test_connect():
    global thread
    with thread_lock:
        if thread is None:
            thread = socketio.start_background_task(target=update_counter)

@socketio.on('disconnect', namespace='/test')
def test_disconnect():
    print('Client disconnected')

if __name__ == '__main__':
    socketio.run(app, debug=True)


#13. Implement notifications in a Flask app using websockets to notify users of updates.

from flask_socketio import SocketIO, emit
import threading
import time

app = Flask(__name__)
socketio = SocketIO(app)

notifications = []
thread = None
thread_lock = threading.Lock()

def send_notifications():
    while True:
        time.sleep(5)
        with thread_lock:
            if notifications:
                notification = notifications.pop(0)
                socketio.emit('notification', {'message': notification}, namespace='/test')

@app.route('/')
def index():
    return render_template('index_notifications.html')

@socketio.on('connect', namespace='/test')
def test_connect():
    global thread
